time to use regression and see if scores and allows impact the wins

```python
nba['diffPTS'] = nba['PTS'] - nba['oppPTS'] #creatig diff between scored and allowed
nba.head()
```

checking the relationship between diffPTS and wins

```python
plt.scatter(nba['diffPTS'], nba['W'], alpha=0.5)
plt.show()
```

```python
import statsmodels.api as sm
```

```python
x = nba['diffPTS']
y = nba['W']
x_sm = sm.add_constant(x)
winRegg = sm.OLS(y, x_sm).fit()
print(winRegg.summary())
```

equation we got: w = 41 + 0.0326*diffPTS // we want 42 wins which means we need 30.67 difference

trying to see what impacts the points

```python
x1 = nba[['2P','3P','FTA','AST','ORB','DRB','TOV', 'STL', 'BLK']]
y1 = nba['PTS']
```

```python
x1_sm = sm.add_constant(x1)
ptsRegg = sm.OLS(y1, x1_sm).fit()
print(ptsRegg.summary())
```

```python
#first removing TOV
x2 = nba[['2P','3P','FTA','AST','ORB','DRB', 'STL', 'BLK']]
x2_sm = sm.add_constant(x2)
ptsRegg = sm.OLS(y1, x2_sm).fit()
print(ptsRegg.summary())
```

```python
x3 = nba[['2P','3P','FTA','AST','ORB', 'STL', 'BLK']]
x3_sm = sm.add_constant(x3)
ptsRegg = sm.OLS(y1, x3_sm).fit()
print(ptsRegg.summary())
```
