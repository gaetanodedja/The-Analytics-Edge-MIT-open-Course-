```python
bball = pd.read_csv(path+filename)
bball.head()
```

```python
mb = bball[bball["Year"] < 2002]
mb.head()
```

```python
mb.info()
```

```python
mb.describe()
```

```python
mb["RD"] = mb["RS"] - mb["RA"]
```

```python
mb.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(mb["RD"],mb["W"], s=3.5)
plt.xlabel("run diff")
plt.ylabel("wins")
plt.show()
```

```python
import statsmodels.api as sm
```

```python
x = mb["RD"]
y = mb["W"]
```

```python
x_sm = sm.add_constant(x)
winRegg = sm.OLS(y, x_sm).fit()
print(winRegg.summary())
```
